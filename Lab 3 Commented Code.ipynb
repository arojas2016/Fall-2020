```
rule = np.zeros((2,2,2))
```

```
#rule 90
rule [0,0,0] = 0
rule [0,0,1] = 1
rule [0,1,0] = 0
rule [0,1,1] = 1
rule [1,0,0] = 1
rule [1,0,1] = 0
rule [1,1,0] = 1
rule [1,1,1] = 0
```

```
def CA(X):

#defines rows and columns respectively
  for i in range(1,X.shape[0]-1): #row
    for j in range(1,X.shape[1]-1): #col

#defines cells above, to the right, and to the left, respectively
      N = X[i-1,j]
      NW = X[i-1,j-1]
      NE = X[i-1,j+1]

      X[i,j] = rule[NW,N,NE]

  return X
```

```
plot(CA(C))
```
