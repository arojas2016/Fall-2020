##Logistic Map Part 2: Poincare Plots, Bifurcation Diagrams, and Cobweb Plots

```
#imports python number/plotting libraries respectively
import numpy as np 
import matplotlib.pyplot as plt
```

```
#defines the logisitc function
def logistic(r, x):
  return r * x * (1-x)
```

```
#defines n, and generates values of n between 2.5 and 4.0
n = 100
r = np.linspace(2.5, 4.0, n)
```

```
#defines x
x= 1e-5 * np.ones(n)
```

```
#defines number of iterations
iterations = 150
```

```
x = logistic(r, x)
```

```
#creates a matrix from the amount of iterations and n
X = np.zeros((iterations, n))
```

```
#assigns the values of the logistic function, that being 1 to the iterations 
for i in range(1, iterations):
  x = logistic(r,x)
  X[i, :] = x
```

```
#plots the first column
plt.plot(X[:,[1]])
```

```
#plots the 81st column
plt.plot(X[:,80], '.')
```

##Birfurcation Diagrams

```
#defines n, and generates values of n between 2.5 and 4.0
n= 100
r = np.linspace(2.5, 4.0, n)
```

```
#defines x
x = 1e-5 * np.ones(n)
```

```
#defining and ploting the plot respectively
fig, ax1= plt.subplots(1, 1, figsize=(10,10))
ax1.plot(r, x, '.', alpha= .25, markersize= 10)
```

```
#defines iterations
iterations = 150
```

```
#defines x as the output of the logistic function involving r and x 
x = logistic(r, x)
```

```
#defines and plots the new function
fig, ax1= plt.subplots(1, 1, figsize=(10, 10))
ax1.plot(r, x, '.', alpha= .25, markersize= 10)
```

```
#defines and plots the new logistic function from 1 to 150 iterations
fig, ax1 = plt.subplots(1, 1, figsize=(10, 10))
for i in range(1, iterations):
  x = logistic(r, x)
  ax1.plot(r, x, '.b', alpha = (i+1) / iterations)
```

##Cobweb Plots

<img src='https://upload.wikimedia.org/wikipedia/commons/e/ed/LogisticCobwebChaos.gif'>

An animated cobweb diagram of the logistic map $y = r x (1-x)$, showing chaotic behaviour for most values of $r > 3.57$.

```
#defines the intervals between 0 and 1, the plot axis, and subsequently plots the figure.
x = np.linspace(0, 1)
fig, ax = plt.subplots(1, 1)
ax.plot(x, logistic(2, x), 'k')
```

```
#defines the intervals between 0 and 1, the plot axis, and subsequently plots the figure.
def plot_system(r, x0, n):
  t = np.linspace(0, 1)

  fig, ax = plt.subplots(1, 1, figsize=(10, 10))
  ax.plot(t, logistic(r, t), 'k', lw= 2)
  ax.plot([0, 1], [0,1], 'k', lw=2)

  x= x0
  
  #iterates the function
  for i in range(n):
    y = logistic(r, x)

    ax.plot([x, x], [x, y], 'k', lw= 1)
    ax.plot([x, x], [y, y], 'k', lw= 1)
    ax.plot([x], [y], 'ok', ms= 10, alpha=(i + 1) / n)

    x = y

  ax.set_xlim(0, 1)
  ax.set_ylim(0, 1)
  ax.set_title(f'$r={r:.1f}, \, x_0={x0:.1f}$')
```

```
#plots
plot_system(2.6, .1, 10)
```

```
#additional plots with different r and x values
plot_system(3.2, .1, 10)
```

```
plot_system(3.9, .1, 10)
```
